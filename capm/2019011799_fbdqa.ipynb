# 3 计算评价指标

```python
def compute_annual_profit(months, net_value):
    """
    计算年化收益
    """
    annual_profit = 0
    # 交易日数大于0，才计算年化收益
    if months > 0:
        # 计算年数
        years = months / 12
        # 计算年化收益
        annual_profit = pow(net_value, 1 / years) - 1

    # 将年化收益转化为百分数，保留两位小数
    annual_profit = round(annual_profit * 100, 2)

    return annual_profit

def compute_drawdown(net_values):
    """
    计算最大回撤
    :param net_values: 净值列表
    """
    # 最大回撤初始值设为0
    max_drawdown = 0
    index = 0
    # 双层循环找出最大回撤
    for net_value in net_values:
        # 计算从当前开始直到结束，和当前净值相比的最大回撤
        for sub_net_value in net_values[index:]:
            # 计算回撤
            drawdown = 1 - sub_net_value / net_value
            # 如果当前的回撤大于已经计算的最大回撤，则当前回撤作为最大回撤
            if drawdown > max_drawdown:
                max_drawdown = drawdown
        index += 1
    return max_drawdown

def compute_sharpe_ratio(net_values):
    """
    计算夏普比率
    :param net_values: 净值列表
    """

    # 总交易日数
    trading_days = len(net_values)
    # 所有收益的DataFrame
    profit_df = pd.DataFrame(columns={'profit'})
    # 收益之后，初始化为第一天的收益
    profit_df.loc[0] = {'profit': round((net_values[0] - 1) * 100, 2)}
    # 计算每天的收益
    for index in range(1, trading_days):
        # 计算每日的收益变化
        profit = (net_values[index] - net_values[index - 1]) / net_values[index - 1]
        profit = round(profit * 100, 2)
        profit_df.loc[index] = {'profit': profit}

    

    # 计算当日收益标准差
    profit_std = pow(profit_df['profit'].var(), 1 / 2)

    # 年化收益
    annual_profit = compute_annual_profit(trading_days, net_values[-1])

    # 夏普比率
    sharpe_ratio = (annual_profit - 3.92) / (profit_std * pow(12, 1 / 2))

    return sharpe_ratio

print('策略（持有）收益:', 
    '{}%'.format(stock1_acc_ret[-1]),
    '{}%'.format(stock2_acc_ret[-1]),
    '{}%'.format(stock3_acc_ret[-1]),
    '{}%'.format(stock4_acc_ret[-1]),
)

print('年化收益:', compute_annual_profit(len(stock1_ret), stock1_acc_ret[-1] + 1),
    compute_annual_profit(len(stock1_ret), stock2_acc_ret[-1] + 1),
    compute_annual_profit(len(stock1_ret), stock3_acc_ret[-1] + 1),
    compute_annual_profit(len(stock1_ret), stock4_acc_ret[-1] + 1))

print('最大回撤:',
    compute_drawdown(stock1_acc_ret + 1) * 100, '%', 
    compute_drawdown(stock2_acc_ret + 1) * 100, '%', 
    compute_drawdown(stock3_acc_ret + 1) * 100, '%', 
    compute_drawdown(stock4_acc_ret + 1) * 100, '%', )

print('夏普利率:', 
    compute_sharpe_ratio(stock1_acc_ret + 1),
    compute_sharpe_ratio(stock2_acc_ret + 1),
    compute_sharpe_ratio(stock3_acc_ret + 1),
    compute_sharpe_ratio(stock4_acc_ret + 1))
```

# 第六题，查看summary可以获取信息

```python
import statsmodels.api as sm

def get_OLS(x, y):
    y = sm.add_constant(y)
    model = sm.OLS(x, y)
    model = model.fit()
    return model

model1 = get_OLS(stock1_over, stock3_over)
model2 = get_OLS(stock2_over, stock3_over)


print(model1.summary())
print(model2.summary())
```

```python


# 总风险
stock1_std = np.std(stock1_over)
stock2_std = np.std(stock2_over)

# 系统性风险
stock1_sys = np.std(model1.predict(sm.add_constant(stock3_ret)))
stock2_sys = np.std(model2.predict(sm.add_constant(stock3_ret)))

# 公司风险
stock1_res = np.std(stock1_over - model1.predict(sm.add_constant(stock3_ret)))
stock2_res = np.std(stock2_over - model2.predict(sm.add_constant(stock3_ret)))

print(stock1_res**2 + stock1_sys**2, stock1_std**2)
print(stock2_res**2 + stock2_sys**2, stock2_std**2)
```
