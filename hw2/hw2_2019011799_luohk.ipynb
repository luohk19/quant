# 题目2

```python
import matplotlib.pyplot as plt
import numpy as np
def simulate(f, verbose=False):
    # 50 次模拟
    total = np.zeros(100)
    for i in range(50):
        curMoney = 100
        series = [curMoney]
        flips = np.random.binomial(1, 0.55, size=99)
        for flip in flips:
            delta = curMoney * f
            curMoney += flip * delta + (flip - 1) * delta
            series.append(curMoney)
        total += np.array(series)
        if verbose:
            plt.plot(range(100), series, color='blue')
    total /= 50
    if verbose:
        plt.plot(range(100), total, color='red', label='f = {}'.format(f))
    else:
        plt.plot(range(100), total, label='f={}'.format(f))
simulate(0.1, True)
plt.show()
```

# 题目3

```python
fs = [0.05, 0.1, 0.25]
for f in fs:
    simulate(f)
plt.legend(loc=2)
plt.show()
```
